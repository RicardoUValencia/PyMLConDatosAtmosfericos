```python
# Importamos las librerias
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
indice1 = pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/indice_2022.csv")
indice2 = pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/indice_2023.csv")
```

```python
indice1.head(25)
```

```python
len(indice1) #tamano
```

```python
#dataframe por dia (promedio) del ozono
#Pero pirmero las separamos, es decir, el ozono del resto
indice = pd.concat([indice1,indice2], axis=0) #Por filas: axis=0
```

```python
indice
```

```python
indice.columns
```

```python
ozono = indice[['Noroeste ozono','Centro ozono','Suroeste ozono', 'Noreste ozono','Sureste ozono']]
```

```python
ozono
```

```python
ozono.info() # Informacion de las columnas
```

```python
#Saber cuantas columnas tienen valores nulos
ozono.isna().sum()
```

```python
# Rellenamos con los valores nulos con la mediana porque es menos sensible a datos atipicos
ozono = ozono.fillna({'Noroeste ozono': ozono['Noroeste ozono'].median(),
              'Noreste ozono': ozono['Noreste ozono'].median(),
              'Centro ozono': ozono['Centro ozono'].median(),
              'Suroeste ozono': ozono['Suroeste ozono'].median(),
              'Sureste ozono': ozono['Sureste ozono'].median(),})
```

```python
ozono
```

```python
#ozono.index = pd.to_datetime(ozono.index, errors='coerce')
#Funcion resample que guarda en un objeto de pandas que se va a mandar traer la informacion de interes
ozono_diario = ozono.resample('D').mean() # La D indica que es por dia
#El resultado es que quedara indexado por dia
```

```python
# Grafica del comportamiento por region
fig, ax = plt.subplots(figsize=(10,5),dpi=200)
ax.plot(ozono['Noroeste ozono'].resample('M').mean(),label='Noroeste',lw=1,color='violet', marker='o',markersize=6)
ax.plot(ozono['Noreste ozono'].resample('M').mean(),label='Noreste',lw=1,color='blue', marker='o',markersize=6)
ax.plot(ozono['Centro ozono'].resample('M').mean(),label='Centro',lw=1,color='orange', marker='o',markersize=6)
ax.plot(ozono['Suroeste ozono'].resample('M').mean(),label='Suroeste',lw=1,color='green', marker='o',markersize=6)
ax.plot(ozono['Sureste ozono'].resample('M').mean(),label='Sureste',lw=1,color='crimson', marker='o',markersize=6)

ax.tick_params(labelsize=20)
ax.tick_params(labelsize=10)
ax.set_xlabel('Tiempo', fontdict={'fontsize':12})
ax.set_ylabel('O3 ($\mu$g/$m^3$)', fontdict = {'fontsize':12});
ax.legend(prop={'size':9})
```

# Tendencia

```python
df = pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/rama_2023_05.csv")
```

```python
!pip install pymannkendall
```

```python
import pymannkendall as mk
```

```python
mk_O3 = mk.original_test(df['O3'],alpha=0.05) #Test de markendall
mk_O3
```

```python
# Traemos de la tupla anterior slope e intercept y guardarlos en a y b
# valores adcionales que serviran para hacer una linea de tendencia con slow e intercept
a = mk_O3.slope
b =mk_O3.intercept
```

```python
def f5(x):
  return x*a+b
```

```python
# Ingresamos un intervalo para que se regresen los valores en y
# Graficamos

# Grafica del comportamiento por region
x = range(0, 3733) #intervalo
fig, ax = plt.subplots(figsize=(10,5),dpi=200)
ax.plot(x,df['O3'],lw=1,color='black');
ax.plot(x,[f5(i) for i in x],color='red',lw=4)
```
