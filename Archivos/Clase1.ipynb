```python
# Importamos las librerias
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Cargamos el dataset
df = pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/rama_2023_05.csv")
```

# Cálculos de media, mediana, varianza y desviación estándar para el CO

```python
# Calculamos la media
df['CO'].mean()
```

```python
# Mediana
df['CO'].median()
```

```python
# Varianza
varianza = df['CO'].var()
print(f'La varianza es: {varianza}')
```

```python
desv_est=np.sqrt(varianza)
print(f'La desviacion estandar es: {desv_est}')
```

# Serie temporal

```python
fig, ax = plt.subplots(figsize=(10,5),dpi=100)
ax.plot(df['O3'], color='black',lw=0.6);
ax.tick_params(labelsize=14)
ax.set_xlabel('Tiempo', fontdict = {'fontsize':14 })
ax.set_ylabel('O3 ($\mu$g/$m^3$)', fontdict = {'fontsize':14 });
ax.axhline(25,color='blue',lw=2,linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5),dpi=100)
ax.plot(df['SO2'], color='black',lw=0.6);
ax.tick_params(labelsize=14)
ax.set_xlabel('Tiempo', fontdict = {'fontsize':14 })
ax.set_ylabel('SO2 ($\mu$g/$m^3$)', fontdict = {'fontsize':14 });
ax.axhline(25,color='green',lw=2,linestyle='--')
plt.show()
```
